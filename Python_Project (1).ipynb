#3) Compare and contrast mutable and immutable objects in Python with examples.
- In Python, objects can be classified into two categories: mutable and immutable. The differences between these two types of objects is essential for writing efficient and effective Python code which are as follows:

Mutable Objects
1. Definition: Mutable objects are those that can be modified after creation.
2. Examples: Lists, dictionaries, sets.
3. Characteristics: Can be changed, extended, or modified after creation.
4. Memory usage: More memory usage due to dynamic allocation.

Mutable objects are useful when we need to modify data after it's been created. For example, we might use a list to store a collection of items that need to be updated frequently.

Immutable Objects
1. Definition: Immutable objects are those that cannot be modified after creation.
2. Examples: Integers, floats, strings, tuples.
3. Characteristics: Cannot be changed, extended, or modified after creation.
4. Memory usage: Less memory usage due to fixed allocation.

Immutable objects are useful when we need to ensure that data remains constant throughout the program. For example, we might use a string to represent a constant value that shouldn't be changed.

##Examples
* Mutable Objects

 * List example
        my_list = [1, 2, 3]
        my_list.append(4)
          print(my_list)  # Output: [1, 2, 3, 4]

  * Dictionary example
        my_dict = {"name": "John", "age": 30}
        my_dict["city"] = "New York"
          print(my_dict)  # Output: {"name": "John", "age": 30, "city": "New York"}

  * Set example
         my_set = {1, 2, 3}
         my_set.add(4)
          print(my_set)  # Output: {1, 2, 3, 4}


* Immutable Objects

  * String example
        my_string = "Hello"
        try:
        my_string[0] = "J"
        except TypeError:
        print("Cannot modify a string")

  * Tuple example
        my_tuple = (1, 2, 3)
        try:
        my_tuple[0] = 4
        except TypeError:
        print("Cannot modify a tuple")

  * Integer example
        my_int = 5
        try:
        my_int += "a"
        except TypeError:
        print("Cannot modify an integer")


Hence, mutable objects can be modified after creation, while immutable objects cannot be modified. Attempting to modify an immutable object results in a TypeError.

Implications for Programming
Understanding the differences between mutable and immutable objects has important implications for programming. When working with mutable objects, we need to be careful not to modify them unintentionally. When working with immutable objects, we can rely on their values remaining constant throughout your program.

By choosing the right data structure for the task provided, developers can write more robust and maintainable code. Whether we're working with mutable or immutable objects, understanding their characteristics and limitations is essential for writing effective Python code.

*   Type casting in Python is the process of converting an object of one data type to another data type. This can be done using various functions and methods provided by Python.

##Types of Type Casting
There are two types of type casting in Python:

###1. Implicit Type Casting:
 This is done automatically by Python when it encounters an expression with mixed data types.

 For example:

a = 5

b= 3.5

print(a + b)      # Output: 8.5


In this example, a is an integer and b is a float. When we add them together, Python automatically converts "a" to a float and performs the addition.
###2. Explicit Type Casting:
 This is done manually by the programmer using various functions and methods.

 For Example:

####Integer to Float

    a = 5
    b = float(a)
    print(b)  # Output: 5.0


####Float to Integer

    a = 5.5
    b = int(a)
    print(b)  # Output: 5


####String to Integer

    a = "5"
    b = int(a)
    print(b)  # Output: 5


####String to Float

    a = "5.5"
    b = float(a)
    print(b)  # Output: 5.5


####List to Tuple

    a = [1, 2, 3]

####Tuple to List

    a = (1, 2, 3)

#6)  How do conditional statements work in Python? Illustrate with examples.
* Conditional statements in Python are used to execute different blocks of code based on certain conditions. They allow the program to make decisions and take different paths depending on the input or situation. There are several types of conditional statements in Python, including if statements, if-else statements, and if-elif-else statements.

##If Statement
  The if statement is used to execute a block of code if a condition is true. The syntax of an if statement is as follows:


if condition:
          # code to be executed


For example:

    x = 5
    if x > 0:
      print("x is positive")


This code will print "x is positive" because the condition x > 0 is true.

##If-Else Statement
The if-else statement is used to execute one block of code if a condition is true and another block if the condition is false. The syntax of an if-else statement is as follows:


if condition:
    # code to be executed if condition is true
else:
    # code to be executed if condition is false


For example:


    age = 25
    if age >= 18:
      print("You are an adult")
    else:
      print("You are a minor")


This code will print "You are an adult" because the condition age >= 18 is true.

##If-Elif-Else Statement
The if-elif-else statement is used to check multiple conditions and execute different blocks of code based on those conditions. The syntax of an if-elif-else statement is as follows:


if condition1:
    # code to be executed if condition1 is true
elif condition2:
    # code to be executed if condition1 is false and condition2 is true
else:
    # code to be executed if all conditions are false


For example:


    grade = 85
    if grade >= 90:
      print("A grade")
    elif grade >= 80:
      print("B grade")
    elif grade >= 70:
      print("C grade")
    else:
      print("Fail")


This code will print "B grade" because the condition grade >= 80 is true.

##Conditional Expressions
Conditional expressions are a concise way to write simple if-else statements. The syntax of a conditional expression is as follows:


value_if_true if condition else value_if_false


For example:

    x = 5
    result = "x is positive" if x > 0 else "x is negative or zero"
    print(result)


This code will print "x is positive" because the condition x > 0 is true.

In conclusion, conditional statements are a fundamental part of programming in Python, and they allow us to write more dynamic and responsive code. By using if statements, if-else statements, and if-elif-else statements, we can control the flow of your program and make decisions based on different conditions.


#7) Describe the different types of loops in Python and their use cases with examples.
*
Loops are a fundamental concept in programming that allow us to execute a block of code repeatedly for a specified number of times or until a certain condition is met. In Python, loops enable us to iterate over sequences, such as lists, tuples, and strings, or perform repetitive tasks.

##Types of Loops:-
###1. For Loop:
 Used to iterate over a sequence (e.g., list, tuple, string) or other iterable objects.
###2. While Loop:
 Used to execute a block of code as long as a certain condition is true.

###1) For Loops:-
For loops are used to iterate over a sequence (such as a list, tuple, or string) or other iterable objects. They allow us to execute a block of code for each item in the sequence.

Example

    fruits = ['apple', 'banana', 'cherry']
    for fruit in fruits:
        print(fruit)


This code will print each fruit in the list.

    Output:- 'apple'
             'banana'
             'cherry'

###2) While Loops:-
While loops are used to execute a block of code as long as a certain condition is true. They are useful when one don't know in advance how many times the loop will need to iterate.

Example

    i = 0
    while i < 5:
      print(i)
      i += 1


This code will print the numbers 0 through 4.

    Output:-   0
               1
               2
               3
               4

###Nested Loops
Nested loops are loops inside loops. They are useful when we need to iterate over multiple sequences or perform complex operations.

Example

    for i in range(3):
      for j in range(2):
        print(f"i: {i}, j: {j}")


This code will print the combinations of i and j.

    Output:-  i: 0, j: 0
              i: 0, j: 1
              i: 1, j: 0
              i: 1, j: 1
              i: 2, j: 0
              i: 2, j: 1

###Loop Control Statements
Loop control statements are used to control the flow of loops. There are two main types:

1. Break: Exits the loop entirely.
2. Continue: Skips to the next iteration of the loop.

Example

    for i in range(5):
      if i == 3:
        break
      print(i)


This code will print the numbers 0 to 2.

Use Cases
1. Iterating over data structures: For loops are ideal for iterating over lists, tuples, and other data structures.
2. Repeating tasks: While loops are useful for repeating tasks until a certain condition is met.
3. Complex operations: Nested loops are useful for performing complex operations that require multiple iterations.

By understanding the different types of loops and their use cases, we can write more efficient and effective code in Python.
